##**CRM Basics and Significance**

### **What is CRM?**

CRM is about building and managing meaningful relationships with customers. It's a comprehensive strategy that goes beyond mere transactions to form deeper, value-driven relationships.

Involves understanding your customers' needs and behaviors to deliver superior customer experiences, thereby fostering customer loyalty and ultimately driving business growth.


### **What are the components of CRM?**

CRM can be broken into 3 main components -
- **Operational CRM** - streamlining and automating marketing, sales and customer service. The aim is to improve the efficiency of these functions and ensure a unified approach to customer interactions.
- **Analytical CRM** - This is focusing on capturing and interpreting and leveraging customer data to gain actionable insights. Through careful analysis, businesses can personalize customer interactions, predict the future behavior and ultimately drive customer loyalty.
- **Collaborative CRM** - This involves sharing customer information among various teams of the company and even sometimes with external stakeholders. The goal is to ensure that everyone involved in servicing customers has a unified, holistic view of the customer.
- **Strategic CRM** - sometimes lumped in with collaborative CRM and provide many of the same features. The difference is that while collaborative CRM focuses on immediate improvements, strategic CRM concentrates on long-term customer engagement. Their main goal is to support customer retention and increase customer loyalty.



### **What is the significance of CRM?**

- Its 5 - 25 times more expensive to acquire a customer than it is to retain an existing one. So by fostering relationships and understanding a users *needs, preferences, behaviours and pain points* we can improve retention thus leading to increased profitability.

- According to Salesforce's State of the Connected Customer report, 88% of customers say the experience a company provides is as important as its products or services. CRM is a key driver for improving customer experiences.
- CRM is also significant in its capacity to surface behavioural signals that inform product and growth decisions. That help improve business decisions and outcomes.

## **Measuring Customer Satisfaction**

### 1. What Are Customer Satisfaction Metrics?
- Metrics used to **quantify how well products, services, or experiences meet customer expectations**
- Provide actionable insights into customer perceptions and experience quality
- A critical component of **CRM decision-making**


### 2. Key Customer Satisfaction Metrics

#### a. Customer Satisfaction Score (CSAT)
- Directly asks customers to rate their satisfaction
- Typically measured on a **sliding or Likert scale**
- Useful for evaluating specific interactions or experiences

#### b. Net Promoter Score (NPS)
- Measures **customer loyalty**
- Asks how likely customers are to recommend a product, service, or company
- Indicates long-term relationship strength

#### c. Customer Effort Score (CES)
- Measures the **ease of customer interaction**
- Often used for:
  - Issue resolution
  - Task completion
- Lower effort = higher satisfaction


### 3. Tools for Measuring Customer Satisfaction

#### a. Surveys
- Primary method for collecting satisfaction data
- Can be delivered via:
  - Email
  - Websites
  - Phone calls

#### b. Feedback Forms
- Allow customers to share opinions, complaints, and suggestions
- Help identify areas for improvement

#### c. Social Media Listening
- Monitors:
  - Brand mentions
  - Industry discussions
- Used to assess **sentiment and satisfaction trends**


### 4. Data Analysis and Action
- Measuring satisfaction alone is not enough
- Real value comes from:
  - Analysing patterns and trends
  - Turning insights into action
- **CRM systems**:
  - Consolidate customer data
  - Enable informed decision-making
- Insights can drive changes to:
  - Products

### 5. Key Takeaway

# CRM Channels

![][image1]

[image1]: <>

CRM channels differ in:

* **Immediacy**
* **Intrusiveness**
* **Personalisation depth**
* **Customer intent**
* **Best-use cases across the customer lifecycle**

Effective CRM strategy uses each channel **for what it does best**, rather than treating them interchangeably.



### 1. Email

#### Behaviour

* **Asynchronous** and non-intrusive
* Customers consume in their own time
* Lower urgency, longer attention span

#### CRM Usage

* Best for **long-form communication**
* Scales well for segmentation and personalisation
* Strong for lifecycle messaging

#### Typical Use Cases

* Onboarding journeys
* Newsletters & content
* Offers & promotions
* Transactional updates
* Re-engagement campaigns

#### Strengths

* High personalisation
* Low cost
* Easy to measure (opens, clicks, conversions)

#### Limitations

* Inbox competition
* Slower response times
* Can suffer from fatigue if overused


### 2. Push Notifications

#### Behaviour

* **Real-time** and highly **interruptive**
* Delivered directly to device
* Requires app installation and opt-in

#### CRM Usage

* Best for **time-sensitive nudges**
* Drives immediate action
* Effective when context-aware (time, location, behaviour)

#### Typical Use Cases

* Abandoned actions
* Reminders
* Limited-time offers
* Product updates

#### Strengths

* High immediacy
* Strong engagement when relevant

#### Limitations

* Easily disabled
* High risk of annoyance
* Short message format


### 3. In-App Messaging

#### Behaviour

* **Contextual and passive**
* Appears while user is actively engaged in the product
* Least intrusive CRM channel

#### CRM Usage

* Best for **guidance and education**
* Supports behaviour change within the product
* Reinforces value at moments of use

#### Typical Use Cases

* Feature discovery
* Onboarding tips
* Upsell prompts
* Error prevention or guidance

#### Strengths

* High relevance
* Strong UX alignment
* No opt-in friction once app is used

#### Limitations

* Only reaches active users
* Limited reach for reactivation


### 4. SMS / WhatsApp

#### Behaviour

* **Immediate and highly intrusive**
* Very high open rates
* Perceived as personal and direct

#### CRM Usage

* Best for **critical or high-value messages**
* Works well for confirmations and urgent alerts
* WhatsApp supports richer, conversational CRM

#### Typical Use Cases

* OTPs & security alerts
* Payment confirmations
* Appointment reminders
* Urgent service updates
* Two-way customer support (WhatsApp)

#### Strengths

* Extremely high visibility
* Fast response
* Strong trust when used sparingly

#### Limitations

* High cost (especially WhatsApp)
* Regulatory constraints
* High risk of customer irritation if overused


### 5. Behavioural Differences Across Channels (Summary)

| Channel      | Immediacy | Intrusiveness | Best For                   | CRM Role          |
| ------------ | --------- | ------------- | -------------------------- | ----------------- |
| Email        | Low       | Low           | Lifecycle & content        | Nurture & retain  |
| Push         | High      | Medium–High   | Time-sensitive nudges      | Activation        |
| In-App       | Medium    | Low           | Contextual guidance        | Engagement        |
| SMS/WhatsApp | Very High | High          | Critical & urgent messages | Trust & assurance |



### 6. Key CRM Principles When Using Channels

* **Match channel to intent**, not just reach
* Use **email for depth**, **push/SMS for urgency**, **in-app for context**
* Overusing high-intrusion channels damages trust
* Best CRM strategies are **orchestrated, not siloed**



### 7. Key Takeaway

Different CRM channels drive **different customer behaviours**.
Effective CRM is about **using the right channel at the right moment** to strengthen long-term relationships, not maximise short-term clicks.
