# Linear Algebra for ML

## Systems of Linear Equations

Linear algebra is one of the most important areas of mathematics for machine learning.

Many machine learning models rely on linear algebra to represent data, express relationships between variables, and compute solutions efficiently.

Understanding systems of linear equations is a foundational step, as these systems appear naturally in many ML problems, including linear regression.

#### **Linear Regression as a Motivating Example**

Linear regression is a supervised machine learning technique where the goal is to learn a relationship between input features and an output target using labelled data. In the simplest case with a single feature, the relationship is modelled as a straight line using the equation \( y = mx + b \), often written in machine learning as \( y = wx + b \), where \( w \) is the weight and \( b \) is the bias. Geometrically, this model corresponds to a line that approximates the trend in the data.

#### **Extending to Multiple Features**

When more than one feature is used, the linear regression model is extended by adding additional weighted inputs. The equation becomes $ y = w_1x_1 + w_2x_2 + \dots + w_nx_n + b $. With two features, this equation represents a plane in three-dimensional space, and with more features it represents a higher-dimensional hyperplane. Although the geometry becomes harder to visualise, the underlying idea remains the same: the output is a weighted sum of the inputs plus a bias term.

#### **From Data Points to Systems of Equations**

Each data point in a dataset produces one linear equation, where the input values and output are known, and the weights and bias are unknown. With many data points, this results in a system of linear equations that must be solved simultaneously. In practice, the goal is not always to find an exact solution, but to find parameter values that best fit all equations as closely as possible.

#### **Linear Algebra Concepts Introduced**

This setup leads directly to key linear algebra concepts such as systems of linear equations, matrix representations, and the distinction between singular and non-singular systems. Whether a system is singular or non-singular determines if a unique solution exists. These ideas form the mathematical foundation for understanding and scaling machine learning models.

#### **Big Picture**

Overall, linear regression provides a practical and intuitive introduction to systems of linear equations. Linear algebra supplies the tools needed to represent these systems efficiently, analyse their properties, and apply them to real-world machine learning problems. The focus of this course is to build a strong mathematical foundation, using machine learning examples to provide intuition and context.

####**Linear Regression as a System of Equations**

In a linear regression problem, each data row contains feature values ($x_1$, $x_2$, $\dots$, $x_n$)  and a target value ($y$\). While the feature values and targets differ for each row, the model parameters (weights \( $w_1$, $\dots$, $w_n$ \) and bias \( $b$ \)) are shared across all rows. The assumption is that a single set of weights and bias can be used to estimate the target for any given set of features.

#### **Vector and Matrix Representation**

Instead of writing the linear model in long form for each feature, it can be written compactly using linear algebra notation. The weights are grouped into a weight vector \( w \), the dataset features form a matrix \( X \), and the targets form a vector \( y \). The model becomes a simple equation that closely resembles the equation of a line, highlighting how vectors and matrices allow complex systems to be expressed cleanly and efficiently.

#### **Linear Algebra in Machine Learning**

Linear algebra focuses on manipulating vectors and matrices to perform powerful computations. In machine learning, linear regression is fundamentally a representation of a system of linear equations. If an exact solution existed, the system could be solved analytically using algebra, provided there are at least as many independent equations as unknown parameters. In practice, machine learning methods solve these systems approximately and iteratively to find the best-fitting solution.

#### **Solving Systems of Linear Equations**

A system of linear equations can be expressed verbally, algebraically, or in matrix form. The unknown variables correspond to model parameters (w), the coefficients correspond to feature values, and the constants correspond to target values. Such systems can be analysed to determine whether they are singular or non-singular, meaning whether they have no solution, a unique solution, or infinitely many solutions.

####**Systems of Equations**



An equation is a mathematical statement that provides information, much like a sentence in natural language. When several equations are considered together, they form a **system of equations**, whose goal is to combine information from each equation in order to determine the values of unknown variables. Just like systems of sentences, systems of equations may contain enough information, too little information, or contradictory information.

A **linear equation** is an equation in which variables appear only multiplied by constants and are combined using addition or subtraction. Linear equations may include a constant term, but they do not allow variables to be multiplied together, raised to powers, or placed inside functions. Examples of linear equations include `a + b = 10`, `2a + 3b = 15`, and `3.4a  48.99b + 2c = 122.5`. Equations involving squares, products of variables, exponentials, logarithms, or trigonometric functions are **non-linear** and are not studied in linear algebra.

A system of linear equations has a **unique solution** when each equation contributes new and independent information. For example, consider the system  
`a + b = 10` and `a + 2b = 12`, where `a` is the price of an apple and `b` is the price of a banana. The difference between the two equations shows that one additional banana costs 2. Substituting this value back into the first equation shows that the apple costs 8. This system is **complete** and **non-singular**, because it leads to exactly one solution.

Systems with more variables can still have a unique solution if enough independent equations are provided. For example, consider the system  
`a + b + c = 10`,  
`a + 2b + c = 15`,  
`a + b + 2c = 12`,  
where `a`, `b`, and `c` represent the prices of an apple, a banana, and a cherry. Comparing the first and second equations shows that a banana costs 5, while comparing the first and third equations shows that a cherry costs 2. Substituting these values into the first equation gives the apple price as 3. Since each variable is determined uniquely, this system is also complete and non-singular.

A system has **infinitely many solutions** when it does not provide enough independent information. For example, the system  
`a + b = 10` and `2a + 2b = 20`  
contains two equations that are mathematically equivalent. The second equation is simply twice the first, so it adds no new information. As a result, any pair of numbers that adds up to 10 is a valid solution. This type of system is called **redundant** and **singular**.

A system has **no solution** when its equations contradict one another. For instance, consider the system  
`a + b = 10` and `2a + 2b = 24`.  
If one apple and one banana cost 10, then two apples and two bananas must cost 20. Since the second equation claims the cost is 24, the information is inconsistent. No values of `a` and `b` can satisfy both equations at once, so the system has no solution. Such systems are described as **contradictory** and **singular**.

The different types of systems of linear equations can be summarised as follows:

| System Type | Description | Number of Solutions |
|------------|-------------|---------------------|
| Complete / Non-Singular | Each equation adds new, independent information | One unique solution |
| Redundant / Singular | Equations repeat the same information | Infinitely many solutions |
| Contradictory / Singular | Equations conflict with each other | No solution |

In conclusion, linear algebra focuses on linear equations because their structure allows information to be combined, compared, and analysed systematically. Understanding whether a system is complete, redundant, or contradictory is essential for determining how much information the system contains and whether a solution can be found.

####**Systems of Equations as Lines and Planes**

Linear equations can be visualised geometrically, which helps make the ideas of solutions, singularity, and non-singularity much clearer. When a linear equation has **two variables**, it can be represented as a **line** in the coordinate plane. When it has **three variables**, it is represented as a **plane** in three-dimensional space. Systems of equations then become collections of lines or planes, and their solutions correspond to how these objects intersect.

#### **Lines and Linear Equations (Two Variables)**

Consider a linear equation such as `a + b = 10`. Each solution to this equation is a pair `(a, b)` whose values add to 10. Examples include `(10, 0)`, `(0, 10)`, `(8, 2)`, `(4, 6)`, and even negative solutions like `(-4, 14)`. When all these points are plotted on a coordinate grid, they lie along a straight line. Every point on this line is a solution to the equation, so the equation and the line represent the same information.

Another equation, such as `a + 2b = 12`, can be visualised in the same way. Points like `(0, 6)`, `(12, 0)`, and `(8, 2)` satisfy the equation, and plotting them produces another straight line. Each linear equation in two variables therefore corresponds to exactly one line in the plane.

#### **Slope and Intercepts**

The **slope** of a line measures how steep the line is and is defined as the ratio of **rise over run**. It tells us how much the vertical value changes when the horizontal value increases by one unit. For the line `a + b = 10`, the slope is `-1`, meaning that for every unit increase in `a`, the value of `b` decreases by one. For the line `a + 2b = 12`, the slope is `-0.5`, meaning that `b` decreases by half a unit for every unit increase in `a`.

The **y-intercept** is the point where the line crosses the vertical axis. For `a + b = 10`, the y-intercept is `10`, while for `a + 2b = 12`, the y-intercept is `6`. These values help distinguish lines visually and explain why different equations produce different geometric objects.

#### **Systems of Lines: Singular and Non-Singular**

When two linear equations are considered together, they form a system that can be visualised as two lines in the same plane. If the system is **non-singular**, the two lines intersect at exactly one point. This point of intersection represents the **unique solution** to the system. For example, the equations `a + b = 10` and `a + 2b = 12` intersect at the point `(8, 2)`. That point is the solution because it satisfies both equations simultaneously.

A system is **singular** if the lines do not intersect at a single point. One possibility is **redundancy**, where the lines overlap exactly. For instance, `a + b = 10` and `2a + 2b = 20` represent the same line. In this case, every point on the line is a solution, so the system has **infinitely many solutions**.

Another singular case occurs when the lines are **parallel** and never intersect. For example, `a + b = 10` and `2a + 2b = 24` produce two distinct parallel lines. Because there is no point that lies on both lines, the system has **no solution**. These equations contradict each other.

#### **Planes and Linear Equations (Three Variables)**

A linear equation with three variables, such as `a + b + c = 1`, is represented by a **plane** in three-dimensional space. Points like `(1, 0, 0)`, `(0, 1, 0)`, and `(0, 0, 1)` satisfy this equation, and any plane passing through these three points contains all solutions to the equation. In general, every linear equation in three variables corresponds to a plane.

When the constant term in a three-variable equation is zero, such as in `3a  5b + 2c = 0`, the plane must pass through the **origin** `(0, 0, 0)`. This is because setting all variables equal to zero satisfies the equation. Any equation of the form `ax + by + cz = 0` therefore represents a plane that goes through the origin.

#### **Systems of Planes: Singular and Non-Singular**

Just as lines can intersect, planes can also intersect in three-dimensional space. A system of three planes is **non-singular** if all planes intersect at exactly one point. For example, the system  
`a + b + c = 0`,  
`a + 2b + c = 0`,  
`a + b + 2c = 0`  
has the unique solution `(0, 0, 0)`. Each plane passes through the origin, and all three intersect at that single point.

A system of planes is **singular** when the intersection is not a single point. One possibility is when planes intersect along a **line**, which leads to **infinitely many solutions**. In this case, every point on the line satisfies all equations in the system.

Another singular case is **redundancy**, where all equations describe the same plane. For example,  
`a + b + c = 0`,  
`2a + 2b + 2c = 0`,  
`3a + 3b + 3c = 0`  
all represent the same plane. Every point on that plane is a solution, so the system again has infinitely many solutions.

#### **Key Geometric Interpretation**

In two variables, solutions to systems correspond to how **lines** intersect: at a point (unique solution), along the same line (infinitely many solutions), or not at all (no solution). In three variables, solutions correspond to how **planes** intersect: at a point (non-singular), along a line (singular with infinitely many solutions), or as overlapping planes (singular and redundant). These geometric interpretations provide an intuitive way to understand singularity, redundancy, and solvability in linear systems.

####**Geometric Notion of Singularity**

Now that systems of linear equations have been visualised as lines in the plane, singularity and non-singularity can be understood in an even simpler and more fundamental way. Previously, three systems were studied: one with a unique solution (complete and non-singular), one with infinitely many solutions (redundant and singular), and one with no solutions (contradictory and singular). While these labels are useful, the key distinction for the rest of the course is **singular vs non-singular**.

When focusing only on singularity, systems 2 and 3 are actually very similar. In both cases, the lines are **parallel**, which makes them fundamentally different from system 1, where the lines are **not parallel** and therefore intersect at a unique point. The difference between systems 2 and 3 lies only in spacing: in system 2 the lines overlap exactly, while in system 3 they are parallel but separated. This observation suggests that singular systems can be grouped together based on parallelism, regardless of whether they overlap or not.

A powerful simplification comes from looking at the **constants** in the equations, meaning the numbers that are not attached to variables. For example, in the systems studied earlier, these constants were values such as 10, 12, 20, or 24. If all constants in each system are set to **zero**, the essential structure of the system is preserved while the equations become easier to analyse.

When the constants are turned into zero, all resulting equations automatically have the point `(0, 0)` as a solution. This means every line in the system must pass through the origin. Geometrically, the systems now consist of lines through `(0, 0)`, which makes their relationships easier to see. Importantly, this transformation does **not** change whether a system is singular or non-singular.

After this simplification, the first system still consists of two non-parallel lines intersecting at a single point, so it remains **non-singular** with a unique solution. The second system still consists of two identical lines, so it remains **singular** with infinitely many solutions. The third system, which originally had no solution due to parallel but distinct lines, now becomes a system of identical lines and therefore has infinitely many solutions. Even though it changes from contradictory to redundant, it remains **singular**, which is the key property of interest.

The crucial conclusion is that **constants do not matter when determining singularity or non-singularity**. Whether lines intersect uniquely or are parallel is determined entirely by the coefficients of the variables, not by the constant terms. For this reason, throughout the rest of the course, systems will often be studied in the simplified form where all constants are zero.

As a result, singular and non-singular systems can be understood geometrically as systems of lines that either all pass through the origin and intersect uniquely (non-singular), or pass through the origin while remaining parallel or overlapping (singular). This simplification allows singularity to be analysed more clearly and consistently, without unnecessary complications from constant terms.
