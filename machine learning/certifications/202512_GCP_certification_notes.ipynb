Predictive AI vs Generative AI

Predictive AI
  - Also known as traditional or discriminative AI
  - uses existing data to classify information or predict future outcomes based on historical patterns
  - excels at learning what is already there to make informed decisions.

Generative AI
  - create summaries
  - uncover complex correlations or generate new content.
  - This includesm text, images or video that mirror the style and patterns within the training data.
  - it doesnt just analyse it creates.

Put it simply, predictive AI analyzes and predicts. Generative AI creates new content and takes action.

when to use them is based on your use case. Generally:

Forecasting & predictions -> predictive AI

Content generation -> generative AI


However there no clear line between these two, and sometimes the best approach is to use both
For example you can use the output from a predictive model as input for a generative model.
For example, use predictive AI to forecast customer churn, then use generative AI to power a chatbot that helps your sales team explore these predictions. Or use predictive AI to identify customer segments, then use generative AI to create personalized marketing content for each segment. Ultimately, business outcomes and customer needs should be prioritised and taken into consideration to help maxmise the benefits of both types of AI.

### AI Infrastructure

The AI infrastructure layer is made up of three layers

- Networking and security (base layer) - lays the foundation to support all of Google's infrastructure and applications.
- Compute & Storage - Google Cloud separates or decouples, as it's technically called, compute and storage so they can scale independently based on need.
- Data and AI products (Top Layer) - Includes data and AI products, which enable you to perform tasks to ingest, store, process, and deliver business insights, data pipelines, and ML models. These can ve accomplished without needing to manage and scale the underlying infrastructure.

Google Cloud Compute Options

Google provides multiple compute services balancing control vs convenience.

- Compute Engine
  - High level of control
  - Similar to managing a physical server
  - User manages OS, runtime, and scaling

- Google Kubernetes Engine (GKE
  - Managed container orchestration
  - Control over containerised applications
  - Handles deployment, scaling, and orchestration

- Cloud Run
  - Serverless platform
  - No infrastructure management
  - Automatically scales based on demand

Where does compute power come from?

Compute power originates from computer chips:

- CPUs – Central Processing Units - general-purpose processing
- GPUs – Graphics Processing Unit - parallel processing, commonly used for ML

However, CPUs and GPUs do not scale efficiently for rapidly growing AI workloads. As a result, Google introduced TPUs in 2016.

Tensor Processing Units (TPUs) are designed specifically to accelerate AI and ML workloads

They act as domain-specific hardware, unlike general-purpose CPUs/GPUs. This is because the architecture is tailored to meet the computation needed in AI. workload, and thus it is optimised for:

- Matrix multiplication
- Neural network training and inference
- Higher performance and energy efficiency

As a result, TPUs are faster and more energy efficient for AI/ML, and integrated into Google Cloud products.

Essentially, As AI workloads grow, organisations move from general-purpose compute (CPUs/GPUs) toward specialised hardware like TPUs, supported by scalable cloud platforms such as Compute Engine, GKE, and Cloud Run.
