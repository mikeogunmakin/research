**Where does compute power come from?**

Compute power originates from computer chips:

- CPUs – Central Processing Units - general-purpose processing
- GPUs – Graphics Processing Unit - parallel processing, commonly used for ML

However, CPUs and GPUs do not scale efficiently for rapidly growing AI workloads. As a result, Google introduced TPUs in 2016.

Tensor Processing Units (TPUs) are designed specifically to accelerate AI and ML workloads

They act as domain-specific hardware, unlike general-purpose CPUs/GPUs. This is because the architecture is tailored to meet the computation needed in AI. workload, and thus it is optimised for:

- Matrix multiplication
- Neural network training and inference
- Higher performance and energy efficiency

As a result, TPUs are faster and more energy efficient for AI/ML, and integrated into Google Cloud products.

Essentially, As AI workloads grow, organisations move from general-purpose compute (CPUs/GPUs) toward specialised hardware like TPUs, supported by scalable cloud platforms such as Compute Engine, GKE, and Cloud Run.

**Supervised vs Unsupervised Learning**

Imagine two types of problems. In problem 1, you are asked to classify dogs and cats from a very large set of pictures. You already know the difference between dogs and cats, so you label each picture and pass the labeled pictures to a machine. By learning from the data, in this case, pictures with the answers or labels, supervised learning is being enacted, allowing the machine to tell if a new picture represents a dog or cat in the future.

In problem 2, you were asked to classify breeds of dogs. Unfortunately, this time, you don't know many of them and are not able to label the pictures.
So you send these unlabeled pictures to a machine. In this case, the machine learns from the data without the answers and finds underlying patterns to group the animals. This is an example of unsupervised learning.

Put simply, supervised learning deals with labeled data, is task-driven, and identifies a goal. Unsupervised learning, however, deals with unlabeled data, is data-driven, and identifies a pattern. An easy way to distinguish between the two is that supervised learning provides each data point with a label or an answer, while unsupervised learning does not.

#### **Types of Supervised Learning**

There are two major types of supervised learning.

The first is classification, which predicts a categorical variable such as determining whether a picture shows a cat or a dog. In ML, you use models like a logistic regression model to solve classification problems.

The second type of supervised learning is regression, which predicts a numeric variable like forecasting sales for a product based on its past sales. You use ML models like a linear regression model to solve regression problems.

#### **Types of Unupervised Learning**

There are three major types of unsupervised learning.

The first is clustering, which groups together data points with similar characteristics and assigns them to clusters, like using customer demographics to determine customer segmentation. You use ML models like k-means clustering to solve clustering problems.

The second type is association, which identifies underlying relationships, like a correlation between two products to place them closer together in a grocery store for a promotion. You use association rule techniques and algorithms like a priori to solve association problems.

And the third type of unsupervised learning is dimensionality reduction, which reduces the number of dimensions or features in a dataset to improve the efficiency of a model. For example, combining customer characteristics like age, driving violation history, or car type to create a simplified rule for calculating an insurance quote. You use ML techniques like principal component analysis to solve these problems.

Although BigQuery began as a traditional data warehouse, it has evolved to support the full data-to-AI lifecycle. In addition to analytics, BigQuery now allows users to build, train, evaluate, and use machine learning models directly using SQL through BigQuery ML. This removes the need to move data between systems and significantly simplifies ML workflows.

Building ML models is typically an iterative and resource-intensive process involving data preparation, feature engineering, model selection, parameter tuning, retraining, and deployment. BigQuery ML streamlines this by letting you manage tabular data and execute ML models in one place. It also automatically handles parts of the workflow, such as parameter tuning and certain preprocessing steps.

A machine learning project in BigQuery ML follows clear phases. First, data is extracted, transformed, and loaded (ETL) into BigQuery if it is not already there. Data can be ingested using built-in connectors from other Google products or enriched using SQL joins across datasets.

Next, feature selection and preprocessing are performed using SQL to create the training dataset. BigQuery ML automatically applies some preprocessing techniques, such as one-hot encoding, which converts categorical variables into numeric form required by ML algorithms.

The model is then created directly inside BigQuery using the `CREATE MODEL` SQL command. For example, predicting whether a customer will purchase a product is a supervised classification problem, making logistic regression an appropriate model choice. BigQuery ML supports multiple model types, including linear regression, logistic regression, k-means clustering, time series forecasting, and more advanced models.

```python
%%bigquery

create model
ecommerce.classification

options
  (
    model_type= 'logistic_reg',
    input_label_cols =
    'will_buy_later'
  ) AS
```

Because classification is a supervised learning task, a label column must be specified. This column represents the outcome the model is trained to predict. After training, model performance is assessed using the `ML.EVALUATE` command, which computes evaluation metrics such as accuracy, precision, and recall.

Once the model performs satisfactorily, predictions are generated using `ML.PREDICT`. The output includes predicted values and confidence scores, with predicted labels clearly marked. BigQuery ML also supports MLOps capabilities, helping teams deploy, monitor, and manage models in production.

Users are encouraged to start with simpler models like linear or logistic regression as benchmarks before moving to more complex models such as deep neural networks, which require greater time and compute resources.
