# **A/B Experiment Design - Step by Step Framework**

## **1. Start with the decision**

Before designing anything, define the business decision the test is meant to inform.

Ask:

- What decision will change if Variant B wins?
- What user behaviour are we trying to influence?
- Why do we think this treatment should work?

A good experiment starts with a clear decision, not just “let’s test this”.

##**2. Define the hypothesis**

Write the hypothesis in a structured way:

If we change X for Y users,
then Z metric will improve,
because of some user or behavioural mechanism.

Example:

> If we simplify the checkout page for new users, then purchase completion rate will increase, because fewer fields should reduce friction.



## **3. Define the Experimental Unit**

Choose the level at which randomisation happens.

Common units:

- user
- user-day
- session
- visits
- geography

Pick the unit that matches how the treatment is experienced and how interference can be avoided.

Example:

- A personalised homepage test usually randomises at user level.
- A pricing test may need account or region level.
- Email subject line tests may randomise at recipient level.

## **4. Define the Population**

Specify exactly who is eligible.

Questions to answer:

- Who is included?
- Who is excluded?
- Is this all users or a segment?
- Are we testing on new users, active users, or returning users?

Document:
- inclusion criteria
- exclusion criteria
- rollout constraints
- traffic allocation

## **5. Define Treatment and Control**

Be explicit about:

- what the control sees
- what the treatment sees
- whether there are multiple treatment arms
- whether the experiment is fixed or adaptive

Write down:

- screenshots or feature flags
- copy changes
- logic changes
- timing differences
- channel differences

## **6. Choose Success Metrics**

Use a metric hierarchy.

**6.1 Primary metric**

The single metric that determines success.

Examples:

- conversion rate
- revenue per user
- click-through rate
- retention rate

**6.2 Secondary metrics**

Helpful supporting outcomes.

Examples:

- funnel step completion
- average order value
- time on page

**6.3 Guardrail metrics**

Metrics that ensure the treatment is not causing harm.

Examples:

- unsubscribe rate
- complaint rate
- page load time
- cancellation rate
- support contacts

**6.4 Diagnostic metrics**

Used to understand mechanism, not to declare a winner.

Examples:

- number of form errors
- scroll depth
- button clicks

A strong design has one primary metric, a few secondary metrics, and a small set of guardrails.

## **7. Define Minimum Detectable Effect**

Decide what effect size is worth detecting.

Ask:

- What is the smallest lift that would matter commercially?

- Is that lift realistic?

- Is the traffic volume sufficient?

This matters because tiny effects may be statistically interesting but operationally irrelevant.

## **8. Estimate sample size and duration**

Use:

- baseline metric level

- desired power

- significance level

- minimum detectable effect

- traffic volume

Then convert required sample size into estimated runtime.

Also account for:

- weekday vs weekend behaviour
- seasonality
- campaign activity
- ramp-up periods



> Recommended Tool for Sample Size and Duration https://www.samplesizecalc.com/



## **9. Plan Randomisation**

Make sure assignment is:

- random
- persistent
- unbiased
- auditable

Check:

- users stay in the same bucket throughout the experiment
- allocation is correct, e.g. 50/50
- bot/internal traffic is handled appropriately
- assignment is logged


# **10. Check for interference and contamination**

Ask whether one unit can affect another.

Examples:

- social sharing can contaminate user-level tests

- marketplace tests can affect supply and demand simultaneously

- email tests can be exposed across devices or accounts

If interference exists, consider:

- cluster randomisation
- geo experiments
- switchback testing
- holdout designs

11. Decide analysis rules before launch

Pre-specify:

- primary estimator
- significance threshold
- handling of missing data
- outlier treatment
- segmentation plan
- stopping rules
- whether multiple testing correction is needed (Bonferroni Correction)

Also decide whether analysis is:

- intention-to-treat
- treatment-on-treated


## **12. Run Pre-launch QA**

Before launch, validate:

- treatment exposure works correctly

- tracking fires correctly

- events are attributed properly

- users are assigned properly

- no major sample ratio mismatch

- dashboards reflect expected counts
