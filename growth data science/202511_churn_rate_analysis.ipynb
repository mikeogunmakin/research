### Asymptote

The asymptote is the maximum theoretical number of customers that a business can accumulate at corresponding acquisition and churn rates.

The higher the customer churn, the closer the asymptote pushes down the cumulative customer base to zero. You can calculate the asymptote as follows:
<br>
<br>
$$
\text{Churn Asymptote} = \frac{\text{Acquisition rate per month}}{\text{Churn rate per month}}
$$
<br>
<br>
For example, at the acquisition rate of 1,000 customers per month and the monthly churn of customers of 10% per month, the maximum possible accumulative base of customers will be = 1,000 / 0.1 = 10,000 customers.
