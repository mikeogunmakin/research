# Retention Rates and their Impact on Lifetime Customer Value

link to article: https://mobiledevmemo.com/mobile-game-retention-rates/

Retention is described as the “King” of mobile app metrics because it is the strongest indicator of whether a game or app is genuinely good. While engagement metrics such as session length and sessions per day provide useful context, retention offers the clearest view of how well the product fits into users’ lives. A commonly accepted benchmark for strong mobile game retention is the 40-20-10 profile: 40% of users returning on day one, 20% on day seven, and 10% on day thirty. Although day-one retention varies depending on acquisition quality and targeting, most games follow a similar pattern of roughly 50% drop-off between key milestones. Well-performing games might achieve retention like 50-25-12, while weaker ones might sit closer to 30-15-7. Overall, retention curves differ across genres and markets, but 40-20-10 is a reasonable goal for most apps.

The retention profile also provides a practical estimate of the average user’s lifetime in calendar days, which forms one half of the lifetime customer value (LCV) calculation. When combined with daily ARPU—also measured in calendar days—it becomes possible to estimate the value of an average user. Because of this connection, retention sits at the foundation of understanding long-term revenue potential and acquisition efficiency.

To measure retention effectively, the author suggests two main data architecture approaches. One option is to store each user in a “users” table and record whether they return on specified retention days. Another is to store one record per user per day of activity. Although the second approach requires more storage, it provides greater flexibility, allowing retention to be analysed relative to any event, such as a major feature release. For example, comparing activity on January 1st to returns on January 2nd can reveal day-one retention for a product launch.

Ultimately, retention is considered the King of all metrics because it reveals the app’s fundamental viability. It tells a product manager not only whether the experience is compelling enough to keep users engaged but also how long users are likely to stay—making retention the essential building block for calculating lifetime value and assessing long-term success.
