##**1. Aim**

**Start by clearly outlining the objective of the analysis**. By mapping out the key stages of the funnel, you can uncover opportunities to optimise the user experience, minimise friction, and improve conversion rates, while analysing user behaviour to identify where drop-offs occur enables you to make data-driven decisions that refine each stage and create a smoother, more effective user journey.



> Key Questions:<br>**What is the business objective? <br> What decision will this inform? <br>What does success look like?**


**Example:** Understand where users drop off in the onboarding funnel to improve activation rate.

## **2. Funnel Definition**

A funnel is a way of modelling how users move through a sequence of steps toward a desired outcome (e.g. conversion). **Clearly define steps or touchpoints in the funnel**. It helps to think about the funnel in 3 stages

- **Top of Funnel (TOF)**: This is the awareness stage, where users first learn about your product or service. The goal here is to attract a wide audience and generate interest. Common tactics include content marketing, social media campaigns, and paid advertising.
- **Middle of Funnel (MOF**): In this stage, users are considering your product and evaluating their options. They may engage with your website, read reviews, or compare features with competitors. The focus here is on nurturing leads and providing valuable information to help users make an informed decision.
- **Bottom of Funnel (BOF)**: This is the conversion stage, where users take the desired action, such as making a purchase or signing up for a free trial. The goal is to remove any remaining barriers and make the conversion process as smooth as possible.

Example:<br>
Step 1: Homepage Visit (TOF) <br> Step 2: Browse Widgets (MOF) <br> Step 3: Add to Cart (MOF) <br> Step 4: Start Purchase Process (BOF) <br> Step 5: Complete Purchase (BOF)

> Key Questions:<br>**what are the key touchpoints or steps?<br> Are the steps linear (strictly sequential) or non-linear (can skip or repeat steps)? <br> is the funnel open (enter at any step) or closed (must start at step 1) <br> is there a time window between steps? should complete all steps within the specified period?**




## **3. Other Considerations**

Before analysing the funnel, **it’s essential to consider the data, metrics, segments and business context to ensure your results are valid and actionable**. Funnel analysis can be misleading if events are incorrectly tracked, steps are poorly defined, or metrics are misinterpreted. This step helps you identify potential biases, data issues, and design choices upfront—so your analysis reflects real user behaviour and leads to reliable, decision-ready insights.

### Data

Does this data truly represent the user journey?

> - **Is event tracking complete and accurate?**<br>
>   e.g. Missing frontend events, untracked steps  
> - **Are timestamps reliable and correctly ordered?**  
>   e.g. Step sequence reflects true user journey  
> - **Is user identity consistent across events and devices?**  
>   e.g. Cookie vs logged-in ID, cross-device stitching  
> - **Are events clean and consistently defined?**  
>   e.g. No duplicates, consistent naming across platforms  
> - **Is the data complete and up to date**?  
>   e.g. No ingestion delays, sampling, or partial data  
> - **Is the dataset correctly filtered?**  
>   e.g. Exclude bots, internal users, test data  



Key data quality factors to consider include:
- Accuracy (data correctly reflects real-world values)
- Completeness (no required data is missing)
- Consistency (data is uniform across systems and datasets)
- Timeliness (data is up to date and available when needed)
- Validity (data conforms to defined formats, rules, and ranges)
- Uniqueness (no duplicate records exist)

(mnemonic: “A Company Checks Their Values Unfailingly”)

### Metrics

What metric(s) best captures our objective(s)?

> - **Conversion rate:** The percentage of users who complete the desired action at each stage of the funnel.  
> - **Drop-off rate:** The percentage of users who leave the funnel at each stage without converting.  
> - **Time to convert:** The average time it takes for users to move through the funnel and complete the desired action.  
> - **Customer acquisition cost (CAC):** The total cost of acquiring a new customer, including marketing and sales expenses.  
> - **Average Order Value (AOV):** The average amount spent per customer order.  
> - **Frequency:** how often a user starts/completes the user funnel in a given period.



Some Principles for Designing Better Metrics:
- Drive real business value — Improving the metric should improve the business
- Keep it simple — Easy to calculate and understand
- Be responsive — Quickly reflects changes (prefer leading indicators)
- Prevent gaming — Hard to manipulate or exploit
- Avoid arbitrary thresholds — Use data-driven cut-offs
- Assign clear ownership — One accountable owner
- Minimise noise — Reduce randomness for clearer signals
- Use standards where needed — Align with industry benchmarks

### Measurement

How should the funnel be measured to accurately reflect user behaviour and the analysis objective?

>- Time - How should time constraints be defined for the funnel?
>   - What is the allowed time window for users to complete all steps? (e.g. 1 hour, 7 days, 30 days, experiment duration)
>   - Should users who exceed this window be treated as drop-offs at the last completed step?
>
>- Counting - How should users be counted in the funnel?
>   - Should the funnel count unique users or total occurrences of events?
>   - If a user re-enters and completes the funnel later, should only the first attempt or all attempts be counted?
> - Trend vs snapshot - How should conversion rates be analysed over time?  
>   - Are we analysing a single time period (snapshot) or tracking conversion over time (trend)?  
>   - Should users be grouped into cohorts (e.g. signup date) to understand changes in funnel behaviour?  

Example:
- **Time:** The time defines that the users should complete all steps within the specified period, for example, 1 hour. It means that it should be 1 hour between viewing the home page and making an order to complete all funnel. If they just complete the first and second steps within 1 hour, they are shown as bounced from the second step.

- **Counting:** If you select unique users, the funnel counts every user for one time in the specified time range (for example, it’s 30 days). For instance, you viewed the home page but didn’t complete the following steps within 1 hour. You returned to the same page after 5 days and completed all the steps. But it doesn’t matter for the funnel. The funnel only counts the first time you complete the first step.
