###What is a Funnel?

A funnel is a way of modelling how users move through a sequence of steps toward a desired outcome

###What are the types of funnels?

- Linear vs Non-linear funnels
  
  - Linear Funnels: A strict step-by-step journey where users cannot skip or repeat steps. E.g Onboard journey.

- Open vs Closed funnels:

  - Open funnels: users can enter the funnel at any stage.
  - Closed funnels: users can only enter the funnel at step 1.


### What is a Funnel bottleneck?

A funnel bottleneck is the stage where the most users are lost, typically indicated by the **lowest step conversion rate**.



###What are the stages in a Funnel?

It helps to think about the funnel in 3 stages:

- **Top of Funnel (TOF)**: This is the awareness stage, where users first learn about your product or service. The goal here is to attract a wide audience and generate interest. Common tactics include content marketing, social media campaigns, and paid advertising.
- **Middle of Funnel (MOF**): In this stage, users are considering your product and evaluating their options. They may engage with your website, read reviews, or compare features with competitors. The focus here is on nurturing leads and providing valuable information to help users make an informed decision.
- **Bottom of Funnel (BOF)**: This is the conversion stage, where users take the desired action, such as making a purchase or signing up for a free trial. The goal is to remove any remaining barriers and make the conversion process as smooth as possible.

### How should you identify the most impactful bottleneck in a funnel?

The most impactful bottleneck is the step where improving conversion leads to the greatest increase in final outcomes. This depends on both conversion rate and user volume, since earlier stages may lose more users in absolute terms. Therefore, the true bottleneck is not necessarily the step with the lowest conversion rate.

###What are the key considerations when doing a funnel analysis?

Considerations:
 1. conversion rate and volume - Don’t fall into the trap of looking at step conversion rates alone.
 2. absolute drop-offs - number of users lost, not just rates.
 3. downstream impact -  "If I improve this step, how much does it move the final metric?”
 4. stage context - TOF (awareness), MOF (consideration) and BOF (conversion).
 5. user intent - Not all users are equal. A drop-off might be healthy filtering rather than a problem (e.g. low-intent users leaving early).
 6. data quality - Bad data = wrong conclusions.
 7. segmentation - Overall funnel metrics can hide issues. Check by device, channel, user type, etc.
 8. effort vs impact - RICE framework (Reach, Impact, Confidence and Effort) to guide recomendations
