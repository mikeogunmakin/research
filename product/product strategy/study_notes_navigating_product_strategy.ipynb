# **Navigating Product Strategy as a Product Data Scientist**

[source](https://medium.com/@AnalyticsAtMeta/data-scientists-framework-for-navigating-product-strategy-as-data-leaders-2eb62b20f505)

### What are the two dimensions in a product strategy framework?

- Data availability (Low → High)
- Problem clarity (Broad → Concrete)

### What are the four quadrants in this framework?

- Pioneer (Low Data, Broad Problem)
- Craftsperson (Low Data, Concrete Problem)
- Explorer (High Data, Broad Problem)
- Optimizer (High Data, Concrete Problem)

### What defines the Pioneer quadrant?

Low data + unclear/broad problem space (early-stage, ambiguous)

### What is your main goal as a Data Scientist in the pioneer quadrant?

Narrow the problem space and create structure

### What are the Key actions for Data Scientist in Pioneer?

- Define north star metric (even conceptual)
- Build metrics framework
- Use qualitative + exploratory analysis
- Run lightweight experiments / quasi-experiments
- Identify proxy metrics

### What’s the biggest value a Data Scientist can add in the Pioneer quadrant?

Turning ambiguity into a **focused, solvable problem**

### What defines the Craftsperson quadrant?

Low data + well-defined problem

###What is the main goal as a Data Scientist in the Craftperson quadrant?

Reduce uncertainty through structured data collection

### What are the Key actions as a Data Scientist in Craftsperson quadrant?

- Design targeted data collection
- Build lightweight measurement frameworks
- Use analogous data from similar problems
- Set learning milestones (not outcomes)

###What does “good execution” as a Data Scientist in the Craftperson quadrant?


Designing a system that can validate hypotheses with minimal data

### What defines the Explorer quadrant?

High data + unclear/broad problem

### Main goal as Data Scietists in the Explorer Quadrant?

Structure the problem space using data

### What are the Key actions for a Data Scientist in Explorer?

- Perform segmentation & clustering
- Identify patterns at scale
- Build opportunity sizing models
- Translate insights into business narratives

### What's the biggest value for Data Scientist in Explorer?

Turning too much data into clear strategic opportunities

###What defines the Optimizer quadrant?

High data + well-defined problem

### What is the Main Goal of Data Scientists in the Optimiser quadrant?

Maximise performance through continuous optimisation

###Key actions in Optimizer?

- Deep metric analysis & monitoring
- Use advanced models (e.g. bandits)
- Run continuous experiments
- Build feedback loops

### Biggest risk in the Optimizer quadrant?

Over-optimising short-term metrics at the expense of long-term strategy

### How does data science role change across segments?

- Pioneer → Define problem
- Craftsperson → Design measurement
- Explorer → Find opportunities
- Optimizer → Drive performance
