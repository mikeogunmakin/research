# User Journey Mapping Framework

### 1. Journey Objective

- Purpose: (e.g., improve conversion, reduce churn, optimise onboarding)
- Scope: (e.g., new users, returning users, specific product/feature)
- Success Criteria: (what a “good” journey looks like)

### 2. Journey Definition

- Start Point: (e.g., first visit, app install, email open)
- End Point: (e.g., conversion, churn, retention milestone)
- Journey Type: (linear, funnel, loop, lifecycle)
- Time Window: (e.g., within 7 days, 30 days)

### 3. Key Stages

- Stage 1: (e.g., Awareness)
- Stage 2: (e.g., Consideration)
- Stage 3: (e.g., Conversion)
- Stage 4: (e.g., Retention)
- Stage 5: (e.g., Loyalty / Repeat)

### 4. Events & Touchpoints

- Key Events: (e.g., page_view, sign_up, purchase)
- Channels: (e.g., email, push, paid media, organic)
- Platforms: (web, app, CRM)
- Event Definitions: (ensure consistent tracking)

### 5. User Paths

- Common Paths: (most frequent sequences)
- Entry Points: (where users enter the journey)
- Exit Points: (where users drop off)
- Path Variations: (different routes to conversion)

### 6. Metrics

- Primary Metric: (e.g., conversion rate)
- Stage Metrics: (e.g., drop-off rate per stage)
- Time Metrics: (e.g., time to convert)
- Engagement Metrics: (e.g., sessions, interactions)

### 7. Friction & Drop-offs

- Drop-off Points: (where users leave)
- Friction Drivers: (UX issues, delays, complexity)
- Behavioural Signals: (hesitation, repeated actions)
- Hypotheses: (why users are dropping off)

### 8. Segmentation

- User Segments: (e.g., new vs returning, high vs low value)
- Channel Segments:
- Cohorts: (e.g., signup week/month)
- Behavioural Segments: (based on actions taken)

### 9. Experimentation Opportunities

- Test Ideas: (e.g., simplify onboarding flow)
- Target Stages: (where to intervene)
- Success Metrics:
- Guardrails:

### 10. Data & Tracking

- Data Sources: (event tracking, CRM, analytics tools)
- Tracking Gaps: (missing or unreliable events)
- Data Quality Checks:
- Granularity: (event-level, session-level)

### 11. Ownership & Usage

- Owner: (team/person responsible)
- Update Frequency:
- Key Stakeholders:
- Use Cases: (product, marketing, CRM, analytics)

### 12. Access & Visualisation

- Dashboard / Tool:
- Journey Visualisation: (funnel, Sankey, pathing)
- Data Location:

### 13. Identity & Journey Rules


- User Definition: (what defines a unique user)
- Identity Stitching: (cross-device, anonymous → known)
- Event Ordering: (timestamp logic, session rules)
- Session Definition: (what defines a session)
- Inclusion Criteria: (who enters the journey)
- Re-entry Rules: (can users restart journeys?)
- Attribution Logic: (how touchpoints are credited)

### 14. Bias & Limitations

- Survivorship Bias: (are only completed journeys included?)
- Path Truncation: (missing early/late events)
- Channel Bias: (e.g., over-crediting retargeting)
- Data Loss: (e.g., cookie restrictions, tracking gaps)
