# **Feature Selection**

## **Overview**

**Feature selection** is the process of selecting a subset of relevant features for use in ML model development.

Why should we select feature?

- simple models are easier to interpret and implement.
- shorter training times (also useful when inference is in real time)
- enhanced generalisation by reducing overfitting (less noise)
- reduced risk of data errors during model use.
- variable redundacy (more features more chance of correlation thus increased risk of redundacy).
- high dimensional spaces not optimal for learning (e.g. high dimensions can cause poor performance in tree based methods, so reducing feature space helps improve predictability and performance)

What is a feature selection procedure?

- a feature selection algorithm can be seen as **the combination of a search technique for proposing new feature subsets, along with an evaluation measure which scores the different feature subsets**.
- ideally a feature selection method would search through all the possible subsets of feature combination that can be obtained from a given data set, and find the feature combination that produces the best machine learnign performance.
  - In practice, this is typically not an option due to the computational cost.
  - Additionally, **different subset of features may produce optimal performance for different ML algorithms**. This means there is not only one optimal subset of features, but potentially many subsets of optimal features depending on the ML alogorith that we intend to use.


What are the types of feature selection methods?
- Filter methods
- Wrapper methods
- Embedded methods

What are Filter methods?

- rely on the characteristics of the the data (feature characteristics)
- They select features independently of any ML algorithm. They rely only on the characteristics of the data (of the variables)
- Do not use ML algorithms
- They are usually univariate, which means each feature is ranked individually and independently of the feature space.
  - Because features are ranked independently of other features, the interaction/relationship between variables is not considered in filter methods, and thus they tend to select redundant variables.
- They are multivairtae filter selection methods as well. features of the data set.Therefore, the multivariate scheme is naturally capable of handling redundant features.

- Usually give lower prediction than other methods.

- Well **suited for quick screen and removel of irrelavant and redundant features**. In practice, filter methods can be good for removing constant features, duplicated features and correlated features.

- **Advantages:** model agnostic and fast computation

- **Disadvantages:** tend to ignore the effects of the selected feature subset on the performance of the induction algorithm.In addition, filter methods often evaluate variables individually, and therefore they're also oblivious to the fact that some variables may not
be predictive in isolation, but they are when combined with other variables.

- additionally, a typical filter method consists of 2 steps:

  - rank features according to a certain criteria. Each feature is ranked independently of the feature space.
  - select the highest ranking features.



What are Wrapper methods?

- use predictive ML algorithm to score the feature subset
- In essence, will build a ML algorithm for each feature subset they evaluate, and then select the suset of variables that produces the highest performing algorithm.
- Thus, they will build several ML models at each round of feature selection (computationally intensive, and might potentially not be possible depending on computing resource).
- Usually provide the best performing feature subset for a given ML algorithm. In theory, the optimal feature subset depends on the specific machine learning algorithm.
  - Note this **may not produce the best feature combination for a different ML model**
  - In practice, **if we used tree derived methods to select the features with wrapper methods, say, for example, we use gradient boosted trees to select features, the optimal set of features selected with this algorithm most likely will also produce a good performance for random forest. However, this set of features may not provide the best performance for logistic regression**.
    - Thus, a practical consideration, is the ML model we intend to build to select the features.


- **Advantages:** First, they are able to detect interactions between variables and second, they find, at least in theory, the optimal subset of features for the desired classifier or regressor. Thus, provide better accuracy than filter methods. Also provide provide best performing subset for a predefined classifier.

- **Disadvantage:** can be computationally expensive. Stopping criteria is relatively arbitrary, and thus limits the theoretical benefits of these procedure


- Wrapper procedure, steps:
   - search for a subset of features
   - build a ML model on the selected feature subset
   - evaluate performance
   - repeat

   Key considerations:
   - how to search for the subset of features? (e.g. forward feature, backward feature or Exhaustive)
   - How to stop the search?


how to search for subset of features?
- Foward feature selection -> adds 1 feature at a time. start with no features ans add features iteratively until ML performancce improves and predefined criteria is met.
- Backward feature selection -> removes 1 feature at a time. removes least significant feature until pre-defined crieteria is met
- Exhaustive feature search -> searches across all possible feature combinations. uses greedy algorithms to find the best feature. can be impractical if feature space is huge


wrapper method stopping criteria?
- performance increase - when performance of the machine learning model does not increase beyond a certain threshold for forward selection. Threshold can be somewhat arbitrary and needs to be defined by user.

- performance decrease - when the performance of the machine learning model algorithm does not decrease beyond a certain threshold for backwards elimination. Threshold can be somewhat arbitrary and needs to be defined by user

- predefined number of features reached. Also need to be predefined by user.




What are Embedded Methods?

- Perform feature selection as part of the model construction process.
- consider the interaction between features and models.
- They are less computationally expensive than wrapper methods, becuase they fit the ML model only once. Thus, select the feature based on the importance attributed to them by the algorithm. These methods are often the preferred choice for better performing feature subsets.

What are some example methods for each technique?

- Filter -> variance, correlation, univariate selection

- Wrapper -> forward selection, backward selection, exhaustive search
